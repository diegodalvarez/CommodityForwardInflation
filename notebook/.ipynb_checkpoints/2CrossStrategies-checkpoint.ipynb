```python
repo_path  = os.path.abspath(os.path.join(os.getcwd(), ".."))
data_path  = os.path.join(repo_path, "data")
latex_path = os.path.join(repo_path, "LaTex")
```

# Trading OLS Residuals Cross-Sectionally

```python
resid_path = os.path.join(data_path, "OLSResid.parquet")
fut_path   = os.path.join(data_path, "FutPX.parquet")

df_resid = pd.read_parquet(path = resid_path, engine = "pyarrow")
df_fut   = pd.read_parquet(path = fut_path, engine = "pyarrow")
```

```python
df_group = (df_resid.pivot(
    index = ["date", "group", "sample_group", "country"], columns = "security", values = "lag_resid").
    assign(med_val = lambda x: x.median(axis = 1)).
    reset_index().
    melt(id_vars = ["date", "group", "sample_group", "country", "med_val"]).
    dropna().
    assign(leg = lambda x: np.where(x.value <= x.med_val, "lgroup", "ugroup")))
```

```python
df_rtn_wider = (df_fut.pivot(
    index = "date", columns = "security", values = "PX_LAST").
    pct_change())
```

```python
df_lagged = (df_rtn_wider.apply(
    lambda x: x * (0.1 / (x.ewm(span = 100, adjust = False).std().shift() * np.sqrt(252)))).
    reset_index().
    melt(id_vars = "date", value_name = "rtn").
    dropna().
    assign(target = "lagged"))

df_perfect = (df_rtn_wider.apply(
    lambda x: x * (0.1 / (x.ewm(span = 100, adjust = False).std() * np.sqrt(252)))).
    reset_index().
    melt(id_vars = "date", value_name = "rtn").
    dropna().
    assign(target = "perfect"))
```

```python
df_combined = (pd.concat([
    df_lagged, df_perfect]).
    merge(right = df_group, how = "inner", on = ["date", "security"]))
```

```python
df_leg_rtn = (df_combined[
    ["date", "target", "group", "sample_group", "leg", "rtn", "country"]].
    groupby(["date", "target", "group", "sample_group", "leg", "country"]).
    agg("mean").
    reset_index().
    assign(name = lambda x: x.group + " " + x.country))
```

```python
names = df_leg_rtn.name.drop_duplicates().sort_values().to_list()
fig, axes = plt.subplots(ncols = 2, nrows = 2, figsize = (20,10))

for name, ax in zip(names, axes.flatten()): 

    (df_leg_rtn.query(
        "name == @name").
        query("sample_group == 'is'").
        assign(
            leg    = lambda x: np.where(x.leg == "lgroup", "Low Resids", "Upper Resids"),
            target = lambda x: np.where(x.target == "lagged", "Lag", "Perf"),
            name   = lambda x: x.leg + " " + x.target).
        rename(columns = {"name": ""}).
        pivot(index = "date", columns = "", values = "rtn").
        apply(lambda x: np.cumprod(1 + x)).
        plot(
            xlabel = "",
            ax     = ax,
            title  = name.replace("forward_inf", "5y5y").replace("suprise_inf", "Sur. Inf."),
            logy   = True,
            ylabel = "Cuml. Ret. (Log-Scaled)"))

fig.suptitle("Cumulative Returns of each L/S Leg using OLS Residuals (In-Samples) Cross-Sectionally\n10% Volatility Targetting (Lagged & Perfect) From {} to {}".format(
    df_leg_rtn.date.min().date(),
    df_leg_rtn.date.max().date()))
plt.tight_layout()
```

```python
df_is_ls_spread = (df_leg_rtn.query(
    "sample_group == 'is'").
    pivot(index = ["date", "target", "group", "country"], columns = "leg", values = "rtn").
    assign(ls_spread = lambda x: 0.5 * (x.lgroup - x.ugroup)).
    reset_index())
```

```python
(df_is_ls_spread.melt(
    id_vars = ["date", "target", "group", "country"]).
    drop(columns = ["date"]).
    groupby(["target", "group", "country", "leg"]).
    agg(lambda x: x.mean() / x.std() * np.sqrt(252)).
    reset_index().
    rename(columns = {
        "target": "Vol. Target",
        "leg"   : "",
        "group" : "Factor"}).
    replace({
        "lgroup"      : "Lower Resid",
        "ugroup"      : "Upper Resid",
        "ls_spread"   : "Net Port",
        "lagged"      : "Lagged",
        "perfect"     : "Perfect",
        "forward_inf" : "5y5y",
        "surprise_inf": "Sur. Inf."}).
    pivot(index = ["Factor", "country"], columns = ["Vol. Target", ""], values = "value").
    apply(lambda x: np.round(x,3)))
```

```python
df_tmp = (df_is_ls_spread.assign(
    name = lambda x: x.country + " " + x.target))

groups    = df_tmp.group.drop_duplicates().to_list()
fig, axes = plt.subplots(ncols = 2, figsize = (20,6)) 

replacer = {
    "surprise_inf": "Surprise Inflation",
    "forward_inf" : "5y5y Forward Inflation"}

for group, ax in zip(groups, axes.flatten()): 

    (df_tmp.query(
        "group == @group").
        rename(columns = {"name": ""}).
        pivot(index = "date", columns = "", values = "ls_spread").
        apply(lambda x: np.cumprod(1 + x)).
        plot(
            ax     = ax,
            ylabel = "Cuml. Ret. (Log-Scaled)",
            logy   = True,
            title  = replacer[group]))

fig.suptitle("Cumulative Returns of Trading L/S Portfolio (rebalanced daily) with 10% Volatility Targeting\nUsing in-sample OLS Residuals from {} to {}".format(
    df_tmp.date.min().date(),
    df_tmp.date.max().date()))
plt.tight_layout()
```

```python
df_ls_spread = (df_leg_rtn.pivot(
    index = ["date", "target", "group", "sample_group", "country"], columns = "leg", values = "rtn").
    assign(ls_rtn = lambda x: 0.5 * (x.lgroup - x.ugroup)).
    reset_index())
```

```python
(df_ls_spread.drop(
    columns = ["lgroup", "ugroup"]).
    drop(columns = ["date"]).
    groupby(["target", "group", "sample_group", "country"]).
    agg(lambda x: x.mean() / x.std() * np.sqrt(252)).
    reset_index().
    rename(columns = {
        "sample_group": "",
        "target"      : "Vol. Target",
        "group"       : "Factor"}).
    replace({
        "is"          : "In-Sample",
        "os"          : "Out-of-Sample",
        "lagged"      : "Lagged",
        "perfect"     : "Perfect",
        "forward_inf" : "5y5y",
        "surprise_inf": "Sur. Inf."}).
    pivot(index = ["Factor", "country"], columns = ["Vol. Target", ""], values = "ls_rtn").
    apply(lambda x: np.round(x,3)))
```
